3. Find the most frequent value and their indice(s) in the following arrays:

i. x = np.array([1,2,3,4,5,1,2,1,1,1])

ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
max_count = np.bincount(x).argmax()
print("max freq number is :", max_count)
print("present at indices")
print([i for i in range(len(x)) if x[i] == max_count])
print()

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
max_count = np.bincount(y).argmax()
print("max freq number is :", max_count)
print("present at indices")
print([i for i in range(len(y)) if y[i] == max_count])
```

4. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'),

find i. Sum of all elements

ii. Sum of all element’s row-wise

iii. Sum of all element’s column-wise

```python
gfg = np.matrix([[4, 1, 9],
                [12, 3, 1],
                [4, 5, 6]])

print("Total sum:", gfg.sum())
print("row-wise sum:",gfg.sum(axis=1))  
print("col-size sum:",gfg.sum(axis=0))  
```
